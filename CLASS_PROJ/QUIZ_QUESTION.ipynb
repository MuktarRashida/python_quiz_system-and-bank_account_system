```python
class listed_question:
    def __init__(self,q,s):
        self.q = list(q)
        self.s = s
    def listed(self):
        self.s = []
        sm = 0
        for j in self.q:
            for k,v in j.items():
                qui = input("{'k'}:Ans: ")
                if qui == v:
                    print(qui)
                    sm += 10
                    self.s.append(sm)
                    print(self.sm)
                else:
                    print(f"correct ans:{v}")
                    print(f"inputted ans:{qui}")
                    sm -= 0
                    self.s.append(sm)
                    print(self.s)
class quizsession:
    from datetime import datetime
    def __init__(self, aq):
        self.aq = aq
    def times(self):
        for v in range(len(self.aq.s)):
            s = self.aq.s[v] 
            f = s % 100
            name = 'TCA CYCLE QUIZ'
            now = self.datetime.now()
            format = now.strftime("%d/%m/%y")
            return f"{name}_{format}_score:{f}"

aq = listed_question([{"what is genomics":"THE STUDY OF DNA"},{"what is NGS":"A FAST READING TECHNOLOGY"},{"what is microbiome":"MICROORGANISMS LIVING IN BODY"}], 0)
print(aq.listed())
qs = quizsession(aq)
print(qs.times())

```
