```python
class listed_question:
    def __init__(self,q,s):
        self.q = list(q)
        self.s = s
    def listed(self):
        sm = 0
        for j in self.q:
            for k,v in j.items():
                qui = input("{'k'}:Ans: ")
                if qui == v:
                    print(qui)
                    sm += 10
                    print(sm)
                else:
                    print(f"correct ans:{v}")
                    print(f"inputted ans:{qui}")
                    sm -= 5
                    print(sm)
t = listed_question([{"what is genomics":"THE STUDY OF DNA"},{"what is NGS":"A FAST READING TECHNOLOGY"},{"what is microbiome":"MICROORGANISMS LIVING IN BODY"}], 20)
print(t.listed())
```
