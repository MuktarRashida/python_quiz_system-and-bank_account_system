```python
class listed_question:
    def __init__(self,q,s):
        self.q = list(q)
        self.s = s
        self.qu = []
    def listed(self):
        self.s = []
        sm = 0
        for j in self.q:
            for k,v in j.items():
                qui = input("{'k'}:Ans: ")
                if qui == v:
                    self.qu.append(qui)
                    print(self.qu)
                    sm += 10
                    self.s.append(sm)
                    print(self.s)
                else:
                    self.qu.append(qui)
                    print(f"correct ans:{v}")
                    print(f"inputted ans:{qui}")
                    sm -= 0
                    self.s.append(sm)
                    print(self.s)
class quizsession:
    from datetime import datetime
    def __init__(self, aq):
        self.aq = aq
    def times(self):
        for v in range(len(self.aq.s)):
            s = self.aq.s[v] 
            f = s % 100
            name = 'TCA CYCLE QUIZ'
            now = self.datetime.now()
            format = now.strftime("%d/%m/%y")
            return f"{name}_{format}_score:{f}"
import json
class result_tracker:
    def __init__(self,aq):
        self.aq = aq
    def ques(self):
        ans = []
        for n in self.aq.q:
            for v in n.values():
                ans.append(v)
        print(f"correct ans:{ans}")
        print(f"inputted ans:{self.aq.qu}")
    def history(self):
        with open("quiz.json",'w') as f:
            da = self.aq.q
            saa = json.dump(da,f)
            print(saa)
    def hist(self):
        with open(r"C:\Users\user\quiz.json",'r') as f1:
            data = json.load(f1)
            print(data)
    


aq = listed_question([{"what is genomics":"THE STUDY OF DNA"},{"what is NGS":"A FAST READING TECHNOLOGY"},{"what is microbiome":"MICROORGANISMS LIVING IN BODY"}], 0)
(aq.listed())
qs = quizsession(aq)
print(qs.times())
rt = result_tracker(aq)
print(rt.ques())
print(rt.history())
print(rt.hist())


```
