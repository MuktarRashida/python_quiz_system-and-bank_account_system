```python
class bankacc:
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance
    def create_acc(self):
        acc = {}
        acc.setdefault('acc-name', self.name)
        acc.setdefault('acc_bal', self.balance)
        print(acc)
        for v in acc.items():
            if v:
                return('account successfully created')
ac = bankacc('rashidat', 20)
print(ac.create_acc())
```
