```python
class bankacc:
    def __init__(self,name,balance):
        if balance < 0:
            raise VaiueError('balance cant be negative')
        self.name = name
        self.balance = balance
    def deposit(self, amount):
        if amount > 5:
            self.balance =  self.balance + amount
            return(self.balance)
    def withdrawal(self,amount1):
        if amount1 > 5 and self.balance >= amount1:
            self.balance -= amount1
            return(self.balance)
    def display(self):
        return(f"{self.name}'s balance: {self.balance}")

class savingsacc(bankacc):
    def __init__(self,name,balance,interest):
        super().__init__(name,balance)
        self.interest = interest
    def rate(self):
        interest = self.balance * self.interest 
        self.balance += interest
        return self.balance
        
ac = bankacc('rashidat',20)
print(ac.deposit(500))
print(ac.withdrawal(200))
print(ac.display())
sv = savingsacc('rashidat',20,0.05)
print(sv.rate())
```
