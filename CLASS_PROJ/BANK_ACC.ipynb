```python
class bankacc:
    def __init__(self,name,balance):
        self.name = name
        self.balance = balance
        self.acc = {}
    def create_acc(self):
        self.acc = {}
        self.acc.setdefault('acc-name', self.name)
        self.acc.setdefault('acc_bal', self.balance)
        print(self.acc)
        for v in self.acc.items():
            if v:
                return('account successfully created')
    def deposit(self, amount):
        for v in self.acc.values():
            if v == self.balance:
                if amount > 5:
                    self.balance = v + amount
                    return(self.balance)
                
ac = bankacc('rashidat', 20)
print(ac.create_acc())
print(ac.deposit(500))
```
